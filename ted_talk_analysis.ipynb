```python
from itertools import combinations
from collections import Counter
```

```python
ted['Grouped'] = ted.groupby(['comments'])['event'].transform(lambda x: ','.join(x))
df2 = ted[['comments','Grouped']].drop_duplicates()
df2.reset_index(drop=True, inplace=True)
df2.head()
```

```python
ted['Grouped'] = ted.groupby(['film_month_year','publish_month_year','comments','views'])['event'].transform(lambda x: ','.join(x))
df2 = ted[['film_month_year','publish_month_year','Grouped','comments','views']].drop_duplicates()
df2.reset_index(drop=True, inplace=True)
df2.head()
```

```python
def get_event1(event):
    return event.split(",")[0].strip(" ")
df2['event1'] = df2['Grouped'].apply(lambda x: '{}'.format(get_event1(x)))
df2.head()
```

```python
ted_events = df2[['film_month_year','publish_month_year','event1','comments','views']]
ted_events.head()
```

```python
events = ted_events.nlargest(12,'comments')
event = events.sort_values(by='comments', ascending=True)
event.reset_index(drop=True, inplace=True)
event
```

#  What were the "best" events in TED history to attend?

```python
top_events = df2['event1']
event_count = Counter()

for event in top_events:
    event_count.update(event.split(';'))

best_events = []
counts=[]
for item in event_count.most_common(10):
    best_events.append(item[0])
    counts.append(item[1])
```

```python
best_events
```

```python
counts
```

```python
plt.style.use('dark_background')
plt.bar(best_events,counts,label='events')
plt.xticks(best_events,rotation='vertical', size=10)
plt.ylabel('event counts')
plt.title('Best event history')
plt.tight_layout()
plt.legend()
plt.show()
```
